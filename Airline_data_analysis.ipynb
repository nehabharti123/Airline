```python
# How the number of tickets booked and total amount earned changed with the time
result_df= pd.read_sql_query("SELECT * FROM tickets INNER JOIN bookings ON tickets.book_ref = bookings.book_ref", connection)
result_df['book_date']= pd.to_datetime(result_df['book_date'])
result_df['date']=result_df['book_date'].dt.date
print(result_df)
```

```python
x = result_df.groupby('date')[['date']].count()
plt.figure(figsize=(18,6))
plt.plot(x.index,x['Date'],marker = '^')
plt.xlabel('date',fontsize=20)
plt.ylabel('Number of tickets',fontsize=20)
plt.grid('b')
plt.show()
```

```python

bookings = pd.read_sql_query("SELECT * FROM bookings", connection)
bookings['book_date'] = pd.to_datetime(bookings['book_date'])
bookings['date'] = bookings['book_date'].dt.date
x = bookings.groupby('date').size().reset_index(name='num_tickets')

plt.figure(figsize=(18, 6))
plt.plot(x['date'], x['num_tickets'], marker='^')
plt.xlabel('Date', fontsize=20)
plt.ylabel('Total amount earned', fontsize=20)
plt.grid(True)
plt.show()
```

```python
# calculate the average charges for each aircraft with different fair conditions
query = """
SELECT fare_conditions, aircraft_code, AVG(amount) as average_amount
FROM ticket_flights
JOIN flights ON ticket_flights.flight_id = flights.flight_id
GROUP BY aircraft_code, fare_conditions
"""

df = pd.read_sql_query(query, connection)
df
```

```python
sns.barplot(data=df, x='aircraft_code', y='average_amount', hue='fare_conditions')
plt.show()
```

```python
# for each aircraft,calculate the total revenue per year and the average revenue per ticket
query = """
SELECT aircraft_code, ticket_count, total_revenue, total_revenue/ticket_count as avg_revenue_per_ticket
FROM (
    SELECT aircraft_code, COUNT(*) as ticket_count, SUM(amount) as total_revenue
    FROM ticket_flights
    JOIN flights ON ticket_flights.flight_id = flights.flight_id
    GROUP BY aircraft_code
) as subquery
"""

df = pd.read_sql_query(query, connection)
df
```

```python
# calculate the average occupancy per aircraft

dff = pd.read_sql_query("""select a.aircraft_code, avg(a.seats_count), b.num_seats, avg(a.seats_count)/b.num_seats, avg(a.seats_count)/b.num_seats as occupancy_rate 
from (select aircraft_code, flights.flight_id, count(*) as seats_count from boarding_passes inner join flights on boarding_passes.flight_id = flights.flight_id group by aircraft_code, flights.flight_id) as a 
inner join (select aircraft_code, count(*) as num_seats from seats group by aircraft_code) as b on a.aircraft_code = b.aircraft_code 
group by a.aircraft_code""", connection)
print(dff)


```
